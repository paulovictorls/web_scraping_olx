```python
from urllib.request import urlopen, urlretrieve, Request
from urllib.error import URLError, HTTPError
from bs4 import BeautifulSoup
import pandas as pd
from math import ceil
```

```python
url = 'https://al.olx.com.br/alagoas/autos-e-pecas/carros-vans-e-utilitarios/up-2018-2018-1-0-mpi-take-up-12v-flex-4p-manual-890670180'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'}

try:
    req = Request(url, headers = headers)
    response = urlopen(req)
    html = response.read().decode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')
    print('succesful read')
    
except HTTPError as e:
    print(e.status, e.reason)
    
except URLError as e:
    print(e.reason)
```

```python
# Title
title = soup.find('h1', class_='sc-1q2spfr-0 lcTcEs sc-ifAKCX cmFKIN').get_text()

# ID
id_ad = url[-9:]

# Date and hour
datetime_list = soup.find('span', class_='sc-1oq8jzc-0 jvuXUB sc-ifAKCX fizSrB').get_text().split()
date_ad = datetime_list[2]
hour_ad = datetime_list[-1]

# Description
description = soup.find('span', class_='sc-1sj3nln-1 eOSweo sc-ifAKCX cmFKIN').get_text().replace('\n',' ')

# Vehicle data
items = soup.findAll('span', class_='sc-ifAKCX dCObfG')
items.pop()
data = []
for i in items:
    data.append(i.get_text())

# Vehicle subdata
subitems = soup.findAll('div', class_='duvuxf-0 h3us20-0 hCwZcX') + soup.findAll('div', class_='duvuxf-0 h3us20-0 kjKryV')
subdata = []

for i in subitems:
    a = i.find('span', class_='sc-ifAKCX cmFKIN')
    b = i.find('a')
    
    if a == None:
        subdata.append(b.get_text())
    else:
        subdata.append(a.get_text())
        
# Optionals
opt = soup.find('div', class_='sc-bwzfXH h3us20-0 cNYGOs').findAll('div', class_="duvuxf-0 h3us20-0 jyICCp")
optionals = []
for o in opt:
    optionals.append(o.get_text())
    
# Professional
professional = False
if soup.find('span', class_='sc-16bj9n5-0 IIBHN sc-ifAKCX fizSrB') is not None: 
    professional = True        

# Location
loc = soup.find('div', class_='h3us20-6 govcZZ').findAll('dd', class_='sc-1f2ug0x-1 ljYeKO sc-ifAKCX kaNiaQ')
loc_title = ['CEP', 'City', 'Neighborhood']
location = []
for l in loc:
    location.append(l.get_text())
    
# Get images
img_list = soup.find('div', class_='h3us20-6 fAprjt').findAll('img', class_='image')

count=1
for i in img_list:
    urlretrieve(i.get('src'), './output/img/' + id_ad + '_' + str(count) + '.jpg')
    count+=1
    
# Creating the card

ad_data = {}
ad_data['Title'] = title
ad_data['ID'] = id_ad
ad_data['Description'] = description

ad_data['Date'] = date_ad
ad_data['Hour'] = hour_ad

if professional: ad_data['Professional'] = 'yes'

count = 0
for d in data:
    ad_data[data[count]] = subdata[count]
    count+=1
    
for o in optionals:
    ad_data[o] = 'yes'

count = 0
for l in location:
    ad_data[loc_title[count]] = location[count]
    count+=1
```

```python
from urllib.request import urlopen, urlretrieve, Request
from urllib.error import URLError, HTTPError
from bs4 import BeautifulSoup
import pandas as pd
from math import ceil
```

```python
def url_reader(url) -> object:

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'}

    try:
        req = Request(url, headers = headers)
        response = urlopen(req)
        html = response.read().decode('utf-8')
        soup = BeautifulSoup(html, 'html.parser')
        print('succesful read')

    except HTTPError as e:
        print(e.status, e.reason)

    except URLError as e:
        print(e.reason)
        
    return soup
```

```python
def list_of_ads(soup) -> list:
    
    """
    insert a html document(soup) from a OLX ad pages and returns the ad links
    """
    
    ad_list = soup.findAll('a', class_='fnmrjs-0 fyjObc')

    ad_links = []
    for ad in ad_list:
        ad_links.append((ad.get('href')))
        
    return ad_links
```

```python
# Acquire the number of pages

main_page = 'https://al.olx.com.br/alagoas/maceio/autos-e-pecas/carros-vans-e-utilitarios'
soup = url_reader(main_page)

aux = soup.find('span', class_='sc-1mi5vq6-0 eDXljX sc-ifAKCX fhJlIo').get_text().split()

ad_per_page = int(aux[2])
n_total_ads = int(aux[-2].replace('.', ''))

n_pages = ceil(n_total_ads/ad_per_page)
```

```python
all_ad_data = []

for page_number in range(n_pages):

    soup = url_reader('https://al.olx.com.br/alagoas/maceio/autos-e-pecas/carros-vans-e-utilitarios?o={}'.format(page_number+1))
    ad_links = list_of_ads(soup)
    
    for ad in ad_links:
    
        soup = url_reader(ad)
        
        # Title
        title = soup.find('h1', class_='sc-1q2spfr-0 lcTcEs sc-ifAKCX cmFKIN').get_text()

        # ID
        aux = soup.find('span', class_='sc-16iz3i7-0 qJvUT sc-ifAKCX fizSrB').get_text().split()
        id_ad = aux[-1]        

        # Date and hour
        datetime_list = soup.find('span', class_='sc-1oq8jzc-0 jvuXUB sc-ifAKCX fizSrB').get_text().split()
        date_ad = datetime_list[2]
        hour_ad = datetime_list[-1]

        # Description
        description = soup.find('span', class_='sc-1sj3nln-1 eOSweo sc-ifAKCX cmFKIN').get_text().replace('\n',' ')

        # Vehicle data
        items = soup.findAll('span', class_='sc-ifAKCX dCObfG')
        items.pop()
        data = []
        for i in items:
            data.append(i.get_text())

        # Vehicle subdata
        subitems = soup.findAll('div', class_='duvuxf-0 h3us20-0 hCwZcX') + soup.findAll('div', class_='duvuxf-0 h3us20-0 kjKryV')
        subdata = []

        for i in subitems:
            a = i.find('span', class_='sc-ifAKCX cmFKIN')
            b = i.find('a')

            if a == None:
                subdata.append(b.get_text())
            else:
                subdata.append(a.get_text())

        # Optionals
        if soup.find('div', class_='sc-bwzfXH h3us20-0 cNYGOs') is not None:
            opt = soup.find('div', class_='sc-bwzfXH h3us20-0 cNYGOs').findAll('div', class_="duvuxf-0 h3us20-0 jyICCp")
            optionals = []
            for o in opt:
                optionals.append(o.get_text())

        # Location
        loc = soup.find('div', class_='h3us20-6 govcZZ').findAll('dd', class_='sc-1f2ug0x-1 ljYeKO sc-ifAKCX kaNiaQ')
        loc_title = ['CEP', 'City', 'Neighborhood']
        location = []
        for l in loc:
            location.append(l.get_text())

        # Get images
        img_list = soup.find('div', class_='h3us20-6 fAprjt').findAll('img', class_='image')

        count=1
        for i in img_list:
            urlretrieve(i.get('src'), './output/img/' + id_ad + '_' + str(count) + '.jpg')
            count+=1

            
            
        # Creating the card

        ad_data = {}
        ad_data['Title'] = title
        ad_data['ID'] = id_ad
        ad_data['Description'] = description

        count = 0
        for d in data:
            ad_data[data[count]] = subdata[count]
            count+=1

        if soup.find('div', class_='sc-bwzfXH h3us20-0 cNYGOs') is not None:
            for o in optionals:
                ad_data[o] = 'yes'

        count = 0
        for l in location:
            ad_data[loc_title[count]] = location[count]
            count+=1
            
        all_ad_data.append(ad_data)
```
